```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'results.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
embedding_scores = data.groupby("embedding_model")["score"].mean().reset_index(name="average_score")
llm_scores = data.groupby("llm_model")["score"].mean().reset_index(name="average_score")

plt.figure(figsize=(10, 6))
sns.barplot(x="average_score", y="embedding_model", data=embedding_scores)
plt.title("Average Score by Embedding Model")
plt.xlabel("Average Score")
plt.ylabel("Embedding Model")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="average_score", y="llm_model", data=llm_scores)
plt.title("Average Score by LLM Model")
plt.xlabel("Average Score")
plt.ylabel("LLM Model")
plt.tight_layout()
plt.show()
```
